# Credit Card 


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Loading the creditcard.csv into a pandas dataframe
#names = ['Age', 'operation_year', 'axillary_nodes_detected', 'survival_status'] 
credit_df = pd.read_csv("creditcard.csv") # added the column names
```

```python
# Total data-points and features
print(credit_df.shape)
```

```python
# total column names
print(credit_df.columns)
```

```python
print(credit_df['Class'].value_counts())
```

```python
# Here, I am separating the classes into separate dataframes
# We have two classes here i.e '1' and '2'
credit_positive = credit_df[credit_df['Class'] == 0]
credit_negative = credit_df[credit_df['Class'] == 1]
print(credit_positive['Class'].value_counts())
print(credit_negative['Class'].value_counts())
```

```python
credit_df["Class"] = np.where(credit_df["Class"].astype(int) == 0,
                                       'Positive','Negative')
credit_df['Class'].value_counts()
```

```python
# top 5 rows in the dataset
print(credit_df.head())
```

```python
# bottom five rows in the dataset
print(credit_df.tail())
```

```python
print(credit_df.describe())
```

```python
# I am using two features i.e; Amount and Time to make a scatter plot
credit_df.plot(kind='scatter', x='Time', y='Amount') ;
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(credit_df, hue = 'Class', size = 4).map(plt.scatter, 'Time', 'Amount').add_legend()
plt.show()
```

```python
sns.FacetGrid(credit_df, hue="Class", size=5) \
   .map(sns.distplot, "Amount") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(credit_df, hue="Class", size=5) \
   .map(sns.distplot, "Time") \
   .add_legend();
plt.show();
```

```python
# Positive class
# Cumulative Distribution Function (CDF)
counts, bin_edges = np.histogram(credit_positive['Amount'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.show();
```

```python
# Positive class
# Cumulative Distribution Function (CDF)
counts, bin_edges = np.histogram(credit_negative['Amount'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.show();
```

```python
# Positive class
counts, bin_edges = np.histogram(credit_positive['Amount'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


# Negative class
counts, bin_edges = np.histogram(credit_negative['Amount'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
```
